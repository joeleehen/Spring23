Select the first name, last name, and payment amount for customers who's last name starts with the letter R, ordered by the amount each customer payed.

```python
!psql -c "SELECT c.first_name, c.last_name, p.amount FROM pagila.customer c INNER JOIN pagila.payment p ON c.customer_id = p.customer_id WHERE c.last_name LIKE 'R%' ORDER BY p.amount LIMIT 10;"
```

Select the title, replacement cost, and inventory ID of movies with a replacement cost greater than $9.99, ordered by decreasing replacement cost

```python
!psql -c "SELECT f.title, f.replacement_cost, i.inventory_id FROM pagila.film AS f INNER JOIN pagila.inventory AS i ON f.film_id = i.film_id WHERE f.replacement_cost >= 9.99 ORDER BY f.replacement_cost DESC LIMIT 10;"
```

Select the ID and address of stores whose address doesn't contain the word "Drive", sorted by address alphabetically.

```python
!psql -c "SELECT s.store_id, a.address FROM pagila.store AS s INNER JOIN pagila.address AS a ON s.address_id = a.address_id WHERE a.address NOT LIKE '%Drive' ORDER BY a.address;"
```

Join film, film category, and category tables to match film names with categories, list all films that are action films, sorted alphabetically by title.

```python
!psql -c "SELECT c.name, f.title FROM pagila.film AS f RIGHT OUTER JOIN pagila.film_category AS r ON f.film_id = r.film_id RIGHT OUTER JOIN pagila.category c ON c.category_id = r.category_id WHERE c.name LIKE '%Action' ORDER BY f.title;"
```

Join first and last names, and emails from the staff table with the address, phone, and postal code from the address table. Select entries that have been updated since February 14, 2017 and group alphabetically by staff last name. 

```python
!psql -c "SELECT s.first_name, last_name, email, a.address, phone, postal_code FROM pagila.staff AS s LEFT OUTER JOIN pagila.address AS a ON s.address_id = a.address_id WHERE DATE(a.last_update) > '2017-02-14' ORDER BY s.last_name;"
```

Combine city, postal code, and address from the city and address tables for cities that do not start with A, sort by postal code and show first 20.

```python
!psql -c "SELECT c.city, a.postal_code, address2, district, address FROM pagila.city AS c LEFT OUTER JOIN pagila.address AS a ON a.city_id = c.city_id WHERE c.city NOT LIKE '%A' ORDER BY a.postal_code LIMIT 20;"
```

Select the title of all pairs of movies that released in the same year, excluding duplicates

```python
!psql -c "SELECT f1.title, f2.title, f2.release_year FROM pagila.film AS f1 INNER JOIN pagila.film AS f2 ON f1.film_id != f2.film_id AND f1.release_year = f2.release_year LIMIT 10;"
```
